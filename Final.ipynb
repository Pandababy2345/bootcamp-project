```python
import pandas as pd
df = pd.read_csv("Tourist_Destinations.csv")
df
```

```python
import plotly.express as px
df = px.data.gapminder().query("Best Season").query("Country")
fig = px.pie(df, values='pop', names='Country',
             title='Most Popular Country to visit',
             hover_data=['Annual Visitors (M)'], labels={'Annual':'Annual Visitors (M)'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
import plotly.express as px

# Filter the DataFrame for Germany
germany_df = df.query("Country == 'Germany'")

# Create a bar chart of Annual Visitors (M) by Type for Germany
fig = px.bar(germany_df, x='Type', y='Annual Visitors (M)', title='Annual Visitors by Destination Type in Germany')
fig.show()
```
